# Trying to imporve model accuracy

```python
data['quality'].value_counts()
```

```python
#Makeing a data the the targer yes for good quality >=3 or No  
df_yn=data.copy()
df_yn["good wine"] = ["No" if index < 6 else "Yes" for index in df_yn['quality']]
X = df_yn.drop(["quality"], axis = 1)
y = df_yn["good wine"]
```

```python
#Makeing a data the the targer yes for good quality >=3 or No  
df_yn=data.copy()
df_yn["good wine"] = [0 if index < 6 else 1 for index in df_yn['quality']]
X = df_yn.drop(["quality"], axis = 1)
y = df_yn["good wine"]
```

```python
y.head()
y.value_counts()
```

Normalization

```python
scaler = MinMaxScaler(feature_range=(0, 1))
normal_df = df_yn.copy()
x = normal_df.drop(columns=['quality'])  # Features
normal_df =  normal_df.drop(columns=['quality'])
x
```

```python
y = data['quality']  # Target  y
```

```python
x = scaler.fit_transform(x)
x = pd.DataFrame(x,columns=normal_df.columns)
x.head()
```


Buillding our model 

```python
# Split the data into training and testing sets
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42,shuffle=True)
print(f"x train shape : {x_train.shape}")
print(f"x test shape : {x_test.shape}")
print(f"Y train shape : {y_train.shape}")
print(f"Y test shape : {y_train.shape}")
```

```python
# Create a Random Forest classifier
rf_model = RandomForestClassifier(n_estimators=1000, max_depth=100, random_state=42)  # You can adjust parameters

# Train the model
rf_model.fit(x_train, y_train)
```

```python
# Make predictions
y_pred = rf_model.predict(x_test)

# Calculate accuracy
accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy: {accuracy * 100:.2f}%')

```

```python
y_predict = list(rf_model.predict(x_test))
predicted_df = {'predicted': y_predict, 'original': y_test}
#creating new dataframe
predicted_df=pd.DataFrame(predicted_df)
predicted_df.head()
```

```python
predicted_df['error']= [True if row['original']==row['predicted'] else False for index,row in predicted_df.iterrows()]
predicted_df.head()
```

```python
predicted_df['error'].value_counts()
```

```python
predicted_df=predicted_df.drop(columns="predicted",axis=1)
predicted_df.head()
```

```python
color = ['#3498DB', '#E74C3C']
# Group by 'original' and 'error', then count occurrences and unstack the 'error' column
test = predicted_df.groupby(['original', 'error'])['original'].count().unstack('error').fillna(0)
# Plotting the bar plot with the specified colors
test.plot(kind='bar', stacked=True, color=color)
# Adding title and customizing the appearance
plt.title('Prediction Error Rate According to Category', fontsize=15)
plt.xticks(rotation=0, fontsize=12)
plt.ylim(0)  # Set the lower limit of the y-axis to 0
# Show the plot
plt.show()
```

Trying different ways to improve the accuracy 

```python
# Create a KNN classifier
knn_model = KNeighborsClassifier(n_neighbors=5)  

# Train the model
knn_model.fit(x_train, y_train)
```

```python
# Make predictions
y_pred = knn_model.predict(x_test)

# Calculate accuracy
accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy: {accuracy * 100:.2f}%')
```

```python
y_predict = list(knn_model.predict(x_test))
predicted_df = {'predicted': y_predict, 'original': y_test}
#creating new dataframe
predicted_df=pd.DataFrame(predicted_df)
predicted_df.head()
```

```python
# Create an SVM classifier
svm_model = SVC(kernel='linear')  

# Train the model
svm_model.fit(x_train, y_train)
```

```python
# Make predictions
y_pred = svm_model.predict(x_test)

# Calculate accuracy
accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy: {accuracy * 100:.2f}%')
```

```python
y_predict = list(svm_model.predict(x_test))
predicted_df = {'predicted': y_predict, 'original': y_test}
#creating new dataframe
predicted_df=pd.DataFrame(predicted_df)
predicted_df.head()
```

```python
# Create an linear regression model 
Linear_Regression_model = LinearRegression()  

# Train the model
Linear_Regression_model.fit(x_train, y_train)
```

```python
# Make predictions
predictions = Linear_Regression_model.predict(x_test)

# Round predictions to the nearest integer
rounded_predictions = np.round(predictions)

# Evaluate the model
mse = mean_squared_error(y_test, rounded_predictions)
print(f"Mean Squared Error (with rounding): {mse}")

# Optionally, convert rounded predictions to integers
rounded_predictions = rounded_predictions.astype(int)
```

```python
y_predict = list(rounded_predictions)
predicted_df = {'predicted': y_predict, 'original': y_test}
#creating new dataframe
predicted_df=pd.DataFrame(predicted_df)
predicted_df.head()
```
